```python
# lấy các bộ phim từ 2010 tới 2015, với rating nhỏ hơn 6.0 nhưng lại có doanh thu thuộc top 5% trên toàn bộ dataset.
df[((df['Year'] >= 2010) & (df['Year'] <= 2015))
& (df['Rating'] < 6.0)
& (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(0.95))]
```

1. `df.groupby('Director')[['Rating']]` - The double square brackets `[['Rating']]` return a DataFrame with one column named 'Rating'. This preserves the DataFrame structure with its column names.
2. `df.groupby('Director')['Rating']` - The single square brackets `['Rating']` return a Series object. This only keeps the values and loses the DataFrame structure.

```python
# tìm số rating trung bình mà các đạo diễn đạt được bằng cách gom nhóm các chỉ số Rating của các bộ phim theo Director.
df.groupby('Director')[['Rating']].mean().sort_values(['Rating'], ascending=False)
```

`apply()` functions: Apply functions được dùng khi ta muốn thực thi một hàm nào đó lên các
hàng trong bảng dữ liệu. Sau khi thực thi, kết quả trả về từ hàm chính là giá trị mới của hàng
tương ứng.

```python
def rating_group(rating):
    if rating >= 7.5:
        return 'Good'
    if rating >= 6.0:
        return 'Average'
    return 'Bad'

df['Rating_category'] = df['Rating'].apply(rating_group)
df.head(5)
```

```python
import pandas as pd

ds_path = './data/opsd_germany_daily.csv'
opsd_daily = pd.read_csv(ds_path)

print(opsd_daily.shape)
print(opsd_daily.dtypes)

opsd_daily.head(5)
```

```python
opsd_daily = opsd_daily.set_index('Date')

opsd_daily.head(5)
```

```python
opsd_daily = pd.read_csv('./data/opsd_germany_daily.csv', index_col=0, parse_dates=True)

opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday Name'] = opsd_daily.index.day_name()

print(opsd_daily.shape)
print(opsd_daily.dtypes)
opsd_daily.sample(5, random_state=0)
```

Time-based indexing: Một trong những tính năng nổi trội của pandas khi xử lý dữ liệu time
series là tính năng time-based indexing, liên quan đến việc dùng dates và times để tổ chức và truy
cập dữ liệu (khá giống với Indexing ở phần trước nhưng giá trị lúc này sẽ là ngày tháng năm).
Việc này cho phép ta dùng loc accessor để thực thi. Ví dụ, ta có thể truy cập dữ liệu theo một
khoảng thời gian từ ngày 2014-01-20 đến ngày 2014-01-22:

```python
opsd_daily.loc['2014-01-20':'2014-01-22']
```

Visualizing time series data: Với việc pandas có hỗ trợ trực quan hóa dữ liệu lên đồ thị, phối
hợp với thư viện seaborn ta có thể dễ dàng trực quan hóa được dữ liệu time-series lên đồ thị. Ví
dụ, ta trực quan (plot) dữ liệu cột Consumption như sau:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(rc={'figure.figsize': (11, 4)})
opsd_daily['Consumption'].plot(linewidth=0.5)
```

```python
cols_plot = ['Consumption', 'Wind', 'Solar']

axes = opsd_daily[cols_plot].plot(
    marker='.', # marker type
    alpha=0.5, # transparency
    linestyle='dotted',  # no line,  supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
    figsize=(11, 9), # size of the figure
    subplots=True, # plot multiple plots
)

for ax in axes:
    ax.set_ylabel('Daily Consumption (MWh)')

plt.tight_layout()


```

```python
data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean() # resample: đổi tần suất dữ liệu, W: tuần
opsd_weekly_mean.head()
```

```python
# visualize daily và weekly time series của Solar trong 6 tháng như sau:
start, end = '2017-01', '2017-06'
fig, ax = plt.subplots()
ax.plot(opsd_daily.loc[start:end, 'Solar'], marker='.', linestyle='-', label='Daily')
ax.plot(opsd_weekly_mean.loc[start:end, 'Solar'], marker='o', linestyle='-', label='Weekly')
ax.set_ylabel('Solar Production (MWh)')
ax.legend()
plt.tight_layout()



```
