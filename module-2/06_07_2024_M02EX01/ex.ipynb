## Q10

```python
def maxx(x,y):
    if x >= y:
        return x
    else:
        return y
    
a = np.array([5,7,9,8,6,4,5])
b = np.array([6,3,4,8,9,7,1])

pair_max = np.vectorize(maxx,otypes=[float])
pair_max(a, b)
```

## Q19

```python
avg_newspaper = np.mean(data[:,2])
newspaper_greater_than_avg_idx = np.where(data[:,2] > avg_newspaper)
np.sum(data[newspaper_greater_than_avg_idx,3])
```

## Q20

```python
avg_sales = np.mean(data[:,3])
scores = np.where(data[:,3] > avg_sales, 'Good', np.where(data[:,3] < avg_sales, 'Bad', 'Average'))
scores[7:10]
```

## Q21

```python
avg_sales = np.mean(data[:, 3])
idx_nearest_avg_sales = np.argmin(np.abs(data[:,3] - avg_sales))
scores = np.where(
    (data[:,3] > avg_sales) & (data[:,3] >= data[idx_nearest_avg_sales, 3]),
    'Good',
    np.where(
        data[:,3] < avg_sales,
        'Bad',
        'Average'
    ),
)
scores[7:10]
```
